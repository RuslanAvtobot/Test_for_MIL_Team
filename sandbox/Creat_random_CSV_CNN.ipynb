Let's do 500 metrics

```python
#Download lib
import torch 
import pandas as pd
import numpy as np
```

```python
#Creat function to make int columnes

def creat_column_int(col_name, low, high):
    a,b = 500,1   #dimension of the pytorch tensor to be generated

    col_name = torch.distributions.uniform.Uniform(low,high).sample([a,b]).int()

    return(col_name)
```

```python
#Creat function to make float columnes

def creat_column_float(col_name, low, high):
    a,b = 500,1   #dimension of the pytorch tensor to be generated

    col_name = torch.distributions.uniform.Uniform(low,high).sample([a,b])

    return(col_name)
```

```python
#first colunme will be count of layers from 5 to 15

leys = creat_column_int('leys', 5, 15)
#print tensor
print(leys)
```

```python
#second columne will be size of kernel size from 2x2 as 2, to 5x5 as 5

ker = creat_column_int('ker', 2, 5)
#print tensor
print(ker)
```

```python
#therd columne will be activation leyer - ReLU - 1, Softmax - 2 or Sigmoid - 3

act = creat_column_int('act', 1, 3)
#print tensor
print(act)
```

```python
#fourth columne will be using max pooling - 1 or avereg - 2 

pool = creat_column_int('pool', 1, 2)
#print tensor
print(pool)
```

```python
#and last colimne time of working CNN model, let's take it from 3 to 25 min

w_time = creat_column_float('w_time', 3, 25)
#print tensor
print(w_time)
```

```python
#and now concat all tensers in one and upload it in csv file

CNN_mod = torch.cat((leys, ker, act, pool, w_time), -1)
print(CNN_mod)
```

```python
CNN_mod_np = CNN_mod.numpy()
CNN_mod_df = pd.DataFrame(data = CNN_mod_np, columns=['leys', 'ker', 'act', 'pool', 'w_time'])
CNN_mod_df.to_csv('CNN_mod.csv')
```
