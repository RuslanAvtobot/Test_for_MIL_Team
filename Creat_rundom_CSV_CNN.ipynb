Let's do 500 metrics

```python
#Download lib
import torch 
import pandas as pd
import numpy as np
```

```python
#first colunme will be count of layers from 5 to 15

a,b = 500,1   #dimension of the pytorch tensor to be generated
low,high = 5,15 #range of uniform distribution

leys = torch.distributions.uniform.Uniform(low,high).sample([a,b]).int()
#print tensor
print(leys)
```

```python
#second columne will be size of kernel size from 2x2 as 2, to 5x5 as 5

low,high = 2,5 #range of uniform distribution

ker = torch.distributions.uniform.Uniform(low,high).sample([a,b]).int()
#print tensor
print(ker)
```

```python
#therd columne will be activation leyer - ReLU - 1, Softmax - 2 or Sigmoid - 3

low,high = 1,3 #range of uniform distribution

act = torch.distributions.uniform.Uniform(low,high).sample([a,b]).int()
#print tensor
print(act)
```

```python
#fourth columne will be using max pooling - 1 or avereg - 2 

low,high = 1,2 #range of uniform distribution

pool = torch.distributions.uniform.Uniform(low,high).sample([a,b]).int()
#print tensor
print(pool)
```

```python
#and last colimne time of working CNN model, let's take it from 3 to 25 min

low,high = 3,25 #range of uniform distribution

w_time = torch.distributions.uniform.Uniform(low,high).sample([a,b])
#print tensor
print(w_time)
```

```python
#and now concat all tensers in one and upload it in csv file

CNN_mod = torch.cat((leys, ker, act, pool, w_time), -1)
print(CNN_mod)
```

```python
CNN_mod_np = CNN_mod.numpy()
CNN_mod_df = pd.DataFrame(data = CNN_mod_np, columns=['leys', 'ker', 'act', 'pool', 'w_time'])
CNN_mod_df.to_csv('CNN_mod.csv')
```
